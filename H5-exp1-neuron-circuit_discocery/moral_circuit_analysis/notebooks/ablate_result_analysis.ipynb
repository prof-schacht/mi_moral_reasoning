```python
# Load json file from disk

with open("../results/ablation/gemma_2_9b_it/care/20250123_174923_gemma_2_9b_it_care_moral_vs_neutral_ablation_results.json", "r") as file:
    moral_neutral_pairs = json.load(file)
```

```python
def print_moral_neutral_pairs(moral_neutral_pairs, dim_name):
    print(f"Dimension: {dim_name}")
    print("Statement 0: moral, Statement 1: neutral")
    for position in range(40):
        for statement in range(2):
            print(f"-------------   Original (Position {position}, Statement {statement})  --------------")
            print(moral_neutral_pairs["original_responses"][position][statement])
            print(f"-------------   Ablated (Position {position}, Statement {statement})  --------------")
            print(moral_neutral_pairs["ablated_responses"][position][statement])
```

```python
print_moral_neutral_pairs(moral_neutral_pairs, "care")
```
